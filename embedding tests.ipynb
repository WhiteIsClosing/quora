```python
embedding_base = '/home/lukasz/Dokumenty/DEEP LEARNING/NLP/embeddings'
model = Word2Vec.load_word2vec_format(os.path.join(embedding_base, 'vecs_300.vec'))
```

```python
from functools import reduce
```

```python
ques = pd.concat([df['question1'], df['question2']], axis=0).tolist()
```

```python
not_in_model = set()
for q in ques:
    if not isinstance(q, str):
        continue
    words = q.split()
    for word in words:
        
        if word not in model:
            not_in_model = not_in_model | set([word])
            
    
print(not_in_model)
```

```python
import re
```

```python
clear_non_ascii = re.compile(r'[^a-z0-9]')
spac = re.compile(r'\s{2,}')
def kill_non_ascii(x):
    if not isinstance(x, str):
        return 'UNK'
    return spac.sub(' ', clear_non_ascii.sub(' ', x))
```

```python
kill_non_ascii('d12we32dlfwa.;4wf.w4;')
```

```python
df[['question1', 'question2']] = df[['question1', 'question2']].applymap(kill_non_ascii)
```

```python
df[['question1', 'question2', 'is_duplicate']].to_csv('preprocessed_to_word2vec.csv')
```
